```python
def icd_parser(input_string):
    # split the string by commas and remove any leading/trailing white space
    codes = [c.strip() for c in input_string.split(',')]
    # wrap the codes with double quotes
    codes = ['"' + c + '"' for c in codes]
    # create a list of lists, with each sublist containing 5 codes
    code_lists = [codes[i:i+5] for i in range(0, len(codes), 5)]
    # join each sublist into a single string with comma separation
    code_strings = [', '.join(cl) for cl in code_lists]
    # add comma at the end of each sublist except for the last one
    for i in range(len(code_strings)-1):
        code_strings[i] = code_strings[i] + ','
    # join the strings with line breaks
    output_string = '\n'.join(code_strings)
    return output_string
```
